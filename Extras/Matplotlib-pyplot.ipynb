```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,100)
```

```python
y = np.sin(x)
```

```python
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

Plotting

```python
# sin wave
plt.plot(x,y)
```

```python
# cosine wave
plt.plot(x,z);
```

```python
# adding title x-axis and y-axis labels

plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('sine wave');
```

```python
# parabola

x = np.linspace(-10,10,20)
y = x**2
plt.plot(x,y)
```

```python
plt.plot(x,y,'r+')
```

```python
plt.plot(x,y,'g.')
```

```python
plt.plot(x,y,'rx')
```

```python
x = np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'g-')
plt.plot(x,np.cos(x),'r--')
```

Bar Plot

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['English','French','Spanish','Latin','Hindi']
people = [100,70,150,40,200]
ax.bar(languages,people)
plt.xlabel('Languages')
plt.ylabel('People')
plt.show()
```

Pie Chart

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
Languages = ['English','French','Spanish','Latin','Hindi']
people = [100,70,150,40,200]
ax.pie(people,labels = Languages, autopct = '%0.1f%%')
plt.show()
```

Scatter Plot

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.show()
```

3D Scatter

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20*np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z,cmap="Blues")
plt.show()
```
